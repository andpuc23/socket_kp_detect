```python
import json
```

```python
class ext_image:
    def __init__(self, id_, path):
        self.id = id_
        self.dataset_id = 16
        self.category_ids = []
        self.path = path
        self.width = 1280
        self.height = 720
        self.file_name = path.split('/')[-1]
        self.annotated = False
        self.annotating = []
        self.num_annotations = 0
        self.metadata = {}
        self.deleted = False
        self.milliseconds = 0
        self.events = []
        self.regenerate_thumbnail = False


class ext_annotation:
    def __init__(self, id_, img_id, keypoints):
        self.id = id_
        self.image_id = img_id
        self.category_id = 5
        self.segmentation = [[244.8, 333.6, 244.8, 91.1, 534.2, 91.1, 534.2,333.6]] # dunno what they do
        self.iscrowd = False
        self.isbbox = True
        self.color = '#eeeeee' # some color
        self.metadata = {}
        
        self.keypoints = []
        for kp in keypoints:
            self.keypoints.append(kp['x'])
            self.keypoints.append(kp['y'])
            self.keypoints.append(kp['state'])
        
        
        xmin = sorted(keypoints, key=lambda k:k['x'])[0]['x']
        xmax = sorted(keypoints, key=lambda k:k['x'])[-1]['x']
        ymin = sorted(keypoints, key=lambda k:k['y'])[0]['y']
        ymax = sorted(keypoints, key=lambda k:k['y'])[-1]['y']
        self.bbox = [xmin, ymin, xmax-xmin, ymax-ymin]
        self.area = int((xmax-xmin)* (ymax-ymin))
    
class extended:
    def __init__(self):
        self.categories = [{
            "id":5,
            'name':'Socket',
            'supercategory':'',
            'color':'#b308ed',
            'metadata':{},
            'keypoint_colors':[]
        }]
        self.images = []
        self.annotations = []
        
    def toJSON(self):
        return json.dumps(self, default=lambda o: o.__dict__, 
            sort_keys=True, indent=2)
```
